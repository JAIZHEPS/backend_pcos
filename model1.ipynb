```python

import pandas as pd
import numpy as np

df = pd.read_csv('PCOS_extended_dataset.csv')

for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

print("Missing values after cleaning:")
print(df.isnull().sum().sum())

# Drop rows with the new NaNs (or you can use df.fillna(df.median(), inplace=True))
df.dropna(inplace=True)

print(f"Cleaned Dataset Shape: {df.shape}")
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 10))
# numeric_only=True is a safety catch for newer pandas versions
sns.heatmap(df.corr(numeric_only=True), annot=False, cmap='mako')
plt.title("Feature Correlation Heatmap")
plt.show()
```
